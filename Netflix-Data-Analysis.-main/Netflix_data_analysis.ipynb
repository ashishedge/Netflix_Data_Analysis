```python
netflix['date_added']=pd.to_datetime(netflix['date_added']) # to change date into date and time format as it was in string format.
```

```python
netflix.fillna({'cast':'unavailable','country':'unavailable','director':'unavailable','rating':'unavailable'},inplace=True)
netflix.isna().sum()  # here we are replacing
```

```python
netflix.type.value_counts()       #  gives the count of different categories in a given column.
```

```python
# a countplot usually helps us to plot counts of each different categories.
sns.countplot(x='type',data=netflix)
plt.title('Count vs Type Of Shows')
```

```python
# now we will do Country analysis.
top_10=netflix['country'].value_counts().head(10)
top_10
```

```python
plt.figure(figsize=(14,7))   # here we are plotting the top 10 countries with most produced contents.
sns.countplot(y='country',order=netflix['country'].value_counts().index[0:10],data=netflix)
plt.title('Country wise content on netflix')
```

```python
country_movie=netflix[netflix.type=='Movie']
country_tv_shows=netflix[netflix.type=='TV Show']
```

```python
plt.figure(figsize=(14,7))   # here we are plotting the top 10 countries with most produced contents(only movies).
sns.countplot(y='country',order=netflix['country'].value_counts().index[0:10],data=country_movie)
plt.title('Country wise movie content on netflix')
```

```python
plt.figure(figsize=(14,7))   # here we are plotting the top 10 countries with most produced contents(including only tv shows.)
sns.countplot(y='country',order=netflix['country'].value_counts().index[0:10],data=country_tv_shows)
plt.title('Country wise tv shows content on netflix')
```

```python
netflix.rating.value_counts()
```

```python
plt.figure(figsize=(10,5))  # ploting of rating of shows on netflix vs count (top 10 countries).
sns.countplot(x='rating',order=netflix['rating'].value_counts().index[0:10],data=netflix)
plt.title('Rating of Shows on Netflix vs Count')
```

```python
netflix.release_year.value_counts()[:20]
```

```python
plt.figure(figsize=(10,6))   # ploting of release year of shows on netflix vs count top 20 countries.
sns.countplot(x='release_year',order=netflix['release_year'].value_counts().index[0:20],data=netflix)
plt.title('Release Year of Shows on Netflix vs Count')
```

```python
# now it is for popular genre analysis (top 20).
plt.figure(figsize=(14,7))
sns.countplot(y='listed_in',order=netflix['listed_in'].value_counts().index[0:20],data=netflix)
plt.title('Top 20 Genre vs Count')
```
